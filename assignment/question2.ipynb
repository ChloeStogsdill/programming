```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 50)

y = np.log(x)
z = np.exp(x)
```

```python
plt.scatter(x,y, label ='Natural Log')
plt.scatter(x,z, label='Exponential')
plt.show()
```

```python
plt.scatter(x, y, label='Natural Log', color='blue')
plt.scatter(x, z, label='Exponential', color='red')
plt.xlabel('x-values')
plt.ylabel('Function values')
plt.title('Natural Log and Exponential Functions')
plt.legend(loc='lower right')
plt.show()
```

```python
x = np.linspace(-6.5, 6.5, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.scatter(x,y_sin)
plt.scatter(x,y_cos)
plt.show()
```

```python
plt.plot(x, y_sin, label='Sine', color='blue')
plt.plot(x, y_cos, label='Cosine', color='green')
plt.show()
```

```python
plt.plot(x, y_sin, label='Sine', color='blue')
plt.plot(x, y_cos, label='Cosine', color='green')
plt.xlabel('x-values')
plt.ylabel('Function values')
plt.title('Sine and Cosine Functions')
plt.legend(loc='lower left')
plt.show()
```
